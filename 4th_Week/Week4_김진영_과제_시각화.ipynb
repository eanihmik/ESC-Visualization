```python
import folium
import pandas as pd
import json
```

## 서울시 구별 고령인구 비율 Choropleth

```python
seoul=json.load(open('data/Seoul_Gu.json'))
seoul
```

```python
pop=pd.read_csv('data/pop_senior_202007.csv',encoding='ansi')
pop.head()
```

```python
pop.columns=['time','sido','gu','seniorRatio','senior','total']
```

```python
m=folium.Map(location=[37.566064, 126.938308],
             tiles='cartodbpositron', zoom_start=10)

folium.Choropleth(geo_data=seoul,
                 data=pop,
                 columns=['gu','seniorRatio'],
                 key_on='feature.properties.SIG_KOR_NM',
                 fill_color='BuGn',
                 fill_opacity=0.8,
                 line_opacity=0.2,
                 legend_name='ratio of aged population (%)').add_to(m)

m
```
